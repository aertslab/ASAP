# Gather samples

in bash


```bash
ls ../../cellranger/*/*/outs/filtered_feature_bc_matrix/matrix.mtx.gz
```

```bash
mkdir -p cellranger_matrices
```

```bash
for file in ../../cellranger/*/*/outs/filtered_feature_bc_matrix/matrix.mtx.gz
do 
    dir=${file%/matrix.mtx.gz}
    name=${file#../../cellranger/*/}
    name=${name%/outs/filtered_feature_bc_matrix/matrix.mtx.gz}
    echo $name

    ln -s $dir/matrix.mtx.gz cellranger_matrices/$name.matrix.mtx.gz
    ln -s $dir/features.tsv.gz cellranger_matrices/$name.features.tsv.gz
    ln -s $dir/barcodes.tsv.gz cellranger_matrices/$name.barcodes.tsv.gz
    printf "\n"
done
```

```bash
for file in ../../cellranger/*/*/outs/raw_feature_bc_matrix.h5
do 
    dir=${file%/raw_feature_bc_matrix.h5}
    name=${file#../../cellranger/*/}
    name=${name%/outs/raw_feature_bc_matrix.h5}
    echo $name

    ln -s $dir/raw_feature_bc_matrix.h5 cellranger_matrices/$name.raw_feature_bc_matrix.h5
    printf "\n"
done
```
