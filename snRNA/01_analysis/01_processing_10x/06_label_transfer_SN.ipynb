````python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scarches.dataset.trvae.data_handling import remove_sparsity
import torch
import scvi
import os

import anndata
import scvi
import scanpy as sc
````

````python
sc.settings.set_figure_params(dpi=200, frameon=False)
sc.set_figure_params(dpi=200)
sc.set_figure_params(figsize=(4, 4))
torch.set_printoptions(precision=3, sci_mode=False, edgeitems=7)

scvi.settings.seed = 94705

%matplotlib inline
%config InlineBackend.print_figure_kwargs={'facecolor' : "w"}
%config InlineBackend.figure_format='retina'
````

````python
model_dir_path = '/CZI/Kamath_Macosko/scANVI_Models/' # path from Julie
````

````python
query_adata = sc.read_h5ad(f'./hdf5/20250210_all_10x_data_sn.h5ad')
````

````python
query_adata.layers['counts'] = query_adata.raw.X if query_adata.raw else query_adata.X
````

````python
query_adata
````

````python
query_adata = query_adata[(query_adata.obs.region != 'Cingulate Gyrus') & (query_adata.obs.region != 'Motor Cortex') ] 
````

````python
query_adata
````

````python
query_adata.obs.region.value_counts()
````

````python
# Add batch column
query_adata.obs['batches_scvi'] = query_adata.obs['sample'].values
````

````python
query_adata
````

````python
# Find any procceses locking the GPU, and if it's not this one, kill it - Use carefully on multi GPU jobs
try:
    nvidia_pid = ! nvidia-smi -q | grep 'Process ID' 
    nvidia_pid = nvidia_pid[0].split(':')[1].strip()

    if os.getpid() != int(nvidia_pid):
        print(f'Killing: {nvidia_pid}')
        ! kill {nvidia_pid}
except:
    print("GPU wasn't locked")
    pass
````

````python
if "PCs" in query_adata.varm.keys():
    del query_adata.varm["PCs"]
````

````python
scvi.model.SCVI.prepare_query_anndata(
    query_adata,
    model_dir_path
)
````

````python
vae_q = scvi.model.SCANVI.load_query_data(
    query_adata,
    model_dir_path,
)
````

````python
# check the setup
vae_q.view_anndata_setup(query_adata)
````

````python
vae_q.train(
    max_epochs=100,
    plan_kwargs=dict(weight_decay=0.0),
    check_val_every_n_epoch=10,
)
````

````python
query_adata.obsm["X_scANVI"] = vae_q.get_latent_representation()
query_adata.obs["predictions"] = vae_q.predict()
````

## Save the predictions

````python
import datetime 

date = datetime.datetime.now().strftime('%Y%m%d')

index_data = query_adata.obs.index
column_data = query_adata.obs['predictions']

# Create a DataFrame with the index and column data
df = pd.DataFrame({'index': index_data, 'column_name': column_data})


# Write the DataFrame to a CSV file
df.to_csv(f'scANVI/{date}_celltype_from_scANVI_sn.csv', index=False)
````
