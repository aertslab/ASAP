```python
def check_columns(row):
    if (row['leiden_2.5']=='64'):
        return 'Lamp5_Exc'
    elif (row['leiden_2.5']=='46'):
        return 'neurons_unk'
    else:
        return row['consensus_cell_type']


adata.obs['consensus_cell_type_corrected'] = adata.obs.apply(lambda row: check_columns(row), axis=1)
```

```python
sc.pl.umap(adata, color = 'consensus_cell_type_corrected')
```

```python
adata.obs.consensus_cell_type_corrected.value_counts()
```
