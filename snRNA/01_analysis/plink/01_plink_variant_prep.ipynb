# Plink

To prepare the population BCF for tensorQTL analysis

```bash
IN_BCF=/WGS_chm13_BCFs_merged/WGS_chm13_BCFs.missing_to_ref.norm.remove_overlaps.bcf # change path for input BCF file
OUT_BCF=WGS_chm13_BCFs.missing_to_ref.norm.remove_overlaps
```

```bash
/software/plink2 --bcf $IN_BCF --make-pgen --set-all-var-ids @_#_\$r_\$a --new-id-max-allele-len 100 --not-chr chrX,chrY,chrM --hwe 1e-5 --max-alleles 2 --maf 0.01 --geno 0.05 --mind 0.05 --out $OUT_BCF
```

# SNPs only

```bash
OUT=WGS_chm13.snps_only
```

```bash
/software/plink2 --bcf $IN_BCF --make-pgen  --snps-only  --write-snplist  \
--set-all-var-ids @_#_\$r_\$a --new-id-max-allele-len 100 --output-chr chrM --not-chr chrX,chrY,chrM \
--hwe 1e-5 --max-alleles 2 --maf 0.01 --geno 0.05 --mind 0.05 --out $OUT
```

# INDELS only

```bash
OUT=/plink/WGS_chm13.indels_only
```

```bash
/software/plink2 --bcf $IN_BCF --make-pgen  --exclude WGS_chm13.snps_only.snplist  \
--set-all-var-ids @_#_\$r_\$a --new-id-max-allele-len 100 --output-chr chrM --not-chr chrX,chrY,chrM \
--hwe 1e-5 --max-alleles 2 --maf 0.01 --geno 0.05 --mind 0.05 --out $OUT
```

# Genotype PCA

## SNP

```bash
/software/plink2 --pfile WGS_chm13.snps_only --pca 5 --out WGS_chm13.snps_only_genotype_pca
```

```bash
/software/plink2 --pfile WGS_chm13.indels_only --pca 5 \
--out WGS_chm13.indels_only_genotype_pca
```
