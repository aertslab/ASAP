## Plot DGE results

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42

import os
````

````python
cc_dge = pd.read_csv('dreamlet/2026_fdp_ASAP_cc_all_bulk_PDPD.tsv', sep = '\t')
cc_dge
````

````python
cc_dge = cc_dge[cc_dge['adj.P.Val'] <= 0.05]
cc_dge['region'] = 'CC'
cc_dge
````

````python
sn_dge = pd.read_csv('dreamlet/2026_fdp_ASAP_sn_all_bulk_PDPD.tsv', sep = '\t')
sn_dge
````

````python
sn_dge = sn_dge[~sn_dge['assay'].isin(['neurons_unk'])]
sn_dge
````

````python
sn_dge = sn_dge[sn_dge['adj.P.Val'] <= 0.05]
sn_dge['region'] = 'SN'
sn_dge
````

````python
dge = pd.concat([sn_dge, cc_dge])
dge
````

````python
dge['ct_reg'] = dge['correct_ct'] + ', ' + dge['region']
dge
````

````python
dge.ct_reg.nunique()
````

````python
import numpy as np
sizes = -np.log10(dge['adj.P.Val'])
trans_sizes = 10 + 150 * (sizes - sizes.min()) / (sizes.max() - sizes.min()) # scale for visibility

def pval_to_size(pval):
    logval = -np.log10(pval)
    return 10 + 150 * (logval - sizes.min()) / (sizes.max() - sizes.min())


# Count occurrences per ct_reg
ct_counts = dge['ct_reg'].value_counts().to_dict()
xticks = dge['ct_reg'].unique()
xtick_labels = [f"{x}\nn={ct_counts.get(x, 0)}" for x in xticks]


plt.figure(figsize=(12, 5))
plt.scatter(
    dge['ct_reg'],
    dge['logFC'],
    s=trans_sizes,
    alpha=0.4,
   # jitter = .3,
    color='dodgerblue',
    edgecolor='k'
)

plt.xticks(ticks=np.arange(len(xticks)), labels=xtick_labels, rotation=45, ha = 'right')

# legend
for val in [1e-20, 1e-10, 1e-05, 1e-02]:
    plt.scatter([], [], s=pval_to_size(val),
                color='dodgerblue', edgecolor='k', alpha=0.7,
                label=f"{val}")
plt.legend(title='Adjusted P-Value', loc='best',
    bbox_to_anchor=(1, 0.5), frameon=True)

plt.ylim(-3.2,3.2)
plt.axhline(0, color='gray', linestyle='--', linewidth=1)

plt.xlabel('')
plt.ylabel('logFC')
plt.title('Differential gene expression')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('dreamlet/fdp_dreamlet_0.05fdr.pdf', dpi=300, bbox_inches='tight')
plt.show()
````
