# Genome binning

Notebook to bin the genome based on Cingulate gyrus ATAC peaks

```python
import pandas as pd
```

```python
# read csv
path = 'CC_CT2_consensus_regions.bed' # swap the path to the correct path to the cingulate cortex consensus regions from snATAC analysis (available on Zenodo))
df = pd.read_csv(path, sep  = '\t', header = None)
df = df.rename(columns = {0:'Chromosome',
                         1:'Start',
                         2:'End'})

df = df[['Chromosome','Start','End']]

# check add region_id
df['region_id']= df['Chromosome'] + ':' + df['Start'].astype(str) + '_' + df['End'].astype(str)
# save as tsv
df
```

```python
df.to_csv('CC_CT2_consensus_regions_genome_binning.tsv',sep="\t", index = False)
```
