# 03 mQTL matrix

This notebook combines information from each donor to a single methylation matrix 

```python
import pandas as pd
import pyranges as pr
import os
import datetime
```

```python
names = [x for x in os.listdir('./donor_binned_data/') if x.endswith('.bed.gz')]
```

```python
names.sort()
names = [x.replace('B.bed.gz', '') for x in names]
print(names)
```

```python
len(names)
```

```python
dfs = {}
for name in names:
    print(name)
    df = pd.read_csv('./donor_binned_data/'+name+'B.bed.gz' )
   # dfs.append(df)
    dfs[name] = df
    print("\n")
```

```python
# combine all the dataframes into a single dataframe where rows are different 

result = []

result = pd.concat(
    (df.set_index('region_id')['Fraction_modified'].rename(name) for name, df in dfs.items()),
    axis=1
)
```

```python
result
```

```python
result = result.reset_index()
```

```python
result.shape
```

```python
# remove rows with more than 5 NAs
# result = result.dropna(thresh=len(result.columns) - 5)
# remove rows with any NA
result = result.dropna()
```

```python
result.shape
```

```python
# get chr start and end from region id
result['Chromosome'] = result['region_id'].str.split(":").str[0]
result['Start'] = result['region_id'].str.extract(r':(\d+)_')[0]
result['End'] = result['region_id'].str.split("_").str[1]
```

```python
# reorder dataframe
cols = result.columns.tolist()  
new_order = cols[-3:] + cols[:-3]  # Rearrange: last 3 columns + the rest
result = result[new_order]
```

```python
result
```

```python
# sort results
result_r = pr.PyRanges(result)
result_r = result_r.sort()
result_final = result_r.df
result_final
```

```python
result_final = result_final.rename({"Chromosome":"#chr", "Start":"start","End":"end"}, axis =1)
```

```python
result_final
```

```python
os.chdir(os.getcwd())
out_dir = "tensor_input_files"
os.makedirs(out_dir, exist_ok = True)
```

```python
date = datetime.datetime.now().strftime('%Y%m%d')
result_final.to_csv(f'{out_dir}/{date}_mQTL_atac_v1_phenotypes.noindex.bed.gz',sep="\t", index = False)
```
