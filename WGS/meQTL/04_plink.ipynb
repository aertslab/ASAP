# Plink

To prepare the population BCF for tensorQTL analysis

```bash
IN_BCF=/ASA/analysis/20241004_patient_demultiplexing/WGS_chm13_BCFs_merged/WGS_chm13_BCFs.missing_to_ref.norm.remove_overlaps.bcf # change to the path with the input BCF file as necessary
```

```bash
mkdir -p plink
```

# SNPs only

```bash
OUT=/plink/WGS_chm13.snps_only.sample_subset
LIST=/plink/sample_list.txt
```

```bash
/software/plink2 --bcf $IN_BCF --make-pgen  --snps-only  --write-snplist  \
--set-all-var-ids @_#_\$r_\$a --new-id-max-allele-len 100 --output-chr chrM --not-chr chrX,chrY,chrM \
--hwe 1e-5 --max-alleles 2 --maf 0.01 --keep $LIST --out $OUT
```

# INDELS only

```bash
OUT=/plink/WGS_chm13.indels_only
SNPLIST=/plink/WGS_chm13.snps_only.sample_subset.snplist
```

```bash
/staging/leuven/stg_00096/home/apancik/software/plink2 --bcf $IN_BCF --make-pgen  --exclude $SNPLIST  \
--set-all-var-ids @_#_\$r_\$a --new-id-max-allele-len 100 --output-chr chrM --not-chr chrX,chrY,chrM \
--hwe 1e-5 --max-alleles 2 --maf 0.01 --keep $LIST --out $OUT
```
